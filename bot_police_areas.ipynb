```python
oakland = pd.read_csv('ca_oakland_2020_04_01.csv')
oakland['lat_lng'] = oakland.apply(lambda x: str(x['lat'])+', '+str(x['lng']), axis=1)
locations = oakland['location'].unique()
coords = oakland['lat_lng'].unique()
```

```python
coords_info = {}
try:
    for coord in tqdm(locations[:1]):
        driver.find_element_by_id('address').send_keys(coord)

        driver.find_element_by_class_name('dijitReset').click()
        coords_info[coord] = driver.find_element_by_class_name('contentPane').text.split('\n')[:4]
        
        
        driver.find_element_by_id('address').clear()
        
except Exception as e:
    print('Error message:', str(e))
```
